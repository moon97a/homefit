```python
import zipfile
import io
import os

# 1. 실제 파일명에 포함된 운동 번호 (런지: 113, 114 / 스쿼트: 115 / 플랭크: 126 / 푸쉬업: 127)
# 위 이미지에서 확인한 113(런지)를 포함하여 리스트를 만들었습니다.
target_numbers = ['113', '114', '115', '126', '127']

zip_files = ['맨몸운동_Labeling.zip', '맨몸운동_Labeling_new_220128.zip']
extracted_list = []

for root_zip_name in zip_files:
    if not os.path.exists(root_zip_name):
        print(f"⚠️ {root_zip_name} 파일이 없습니다.")
        continue
        
    print(f"📦 {root_zip_name} 분석 중...")
    
    with zipfile.ZipFile(root_zip_name, 'r') as root_zip:
        for inner_file_name in root_zip.namelist():
            if inner_file_name.endswith('.zip'):
                with root_zip.open(inner_file_name) as inner_zip_data:
                    zip_content = io.BytesIO(inner_zip_data.read())
                    try:
                        with zipfile.ZipFile(zip_content) as inner_zip:
                            # 내부 파일 목록 확인
                            for f in inner_zip.namelist():
                                # 파일명이 '.json'으로 끝나고, 우리 타겟 번호 중 하나를 포함하는지 확인
                                # 예: 'D05-8-113.json'에서 '113'이 있는지 검사
                                if f.endswith('.json') and any(num in f for num in target_numbers):
                                    extracted_list.append({
                                        'root': root_zip_name,
                                        'inner': inner_file_name,
                                        'file': f
                                    })
                    except:
                        continue

print("-" * 30)
print(f"✅ 총 {len(extracted_list)}개의 파일을 찾았습니다!")

if extracted_list:
    print("찾은 파일 예시:", extracted_list[0]['file'])
```

```python
import zipfile
import io
import os

# 저장할 폴더 이름
output_folder = 'extracted_exercises'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

print(f"📂 '{output_folder}' 폴더에 파일 추출을 시작합니다...")

# 아까 찾은 952개의 리스트(extracted_list)를 사용합니다.
for item in extracted_list:
    root_zip_name = item['root']
    inner_zip_name = item['inner']
    json_file_name = item['file']
    
    with zipfile.ZipFile(root_zip_name, 'r') as root_zip:
        with root_zip.open(inner_zip_name) as inner_zip_data:
            zip_content = io.BytesIO(inner_zip_data.read())
            with zipfile.ZipFile(zip_content) as inner_zip:
                # 파일명만 추출해서 저장 (폴더 구조 제외하고 파일명만 깔끔하게)
                simple_name = os.path.basename(json_file_name)
                content = inner_zip.read(json_file_name)
                
                with open(os.path.join(output_folder, simple_name), 'wb') as f:
                    f.write(content)

print(f"✨ 추출 완료! '{output_folder}' 폴더를 확인해보세요.")
```
